```r
install.packages("pacman")
pacman::p_load("tidyverse", "nycflights13", "datasauRus")
```

```r
nycflights13::flights
```

```r
class(flights)
```

```r
tibb <- flights
df <- as.data.frame(tibb)
```

```r
df[df$carrier == "UA", c("dep_delay")]
```

```r
tibb[tibb$carrier == "UA", c("dep_delay")]
```

```r
subset(df, subset = carrier == "UA", select = c('dep_delay', 'arr_delay'))
```

```r
tibb %>%
dplyr::filter(carrier == "UA") %>%
dplyr::select(dep_delay, arr_delay)
```

```r
df$arr_delay[df$carrier == "UA"]
```

```r
plot(
    x = tibb$arr_delay[tibb$carrier == "UA"], 
    y = tibb$dep_delay[tibb$carrier == "UA"], 
    pch = 17, 
    main = "Scatter", 
    xlab = "Arr_Delay", 
    ylab = "Dep_Delay",
    col = "purple"
    )
```

```r
plot(
    x = tibb$arr_delay[tibb$carrier == "UA"], 
    y = tibb$dep_delay[tibb$carrier == "UA"], 
    pch = 17, 
    main = "Scatter", 
    xlab = "Arr_Delay", 
    ylab = "Dep_Delay",
    col = "purple"
    )
points(
    tibb$arr_delay[tibb$carrier == "AA"],
    tibb$dep_delay[tibb$carrier == "AA"],
    pch = 23,
    col = "red"
    )
mtext("Scatter plot for UA and AA - arr_delay vs dep_delay")
```

```r
par(mfrow = c(1,2))
plot(
    x = tibb$arr_delay[tibb$carrier == "UA"], 
    y = tibb$dep_delay[tibb$carrier == "UA"], 
    pch = 17, 
    main = "UA", 
    xlab = "Arr_Delay", 
    ylab = "Dep_Delay",
    col = "purple"
    )
plot(
    x = tibb$arr_delay[tibb$carrier == "AA"],
    y = tibb$dep_delay[tibb$carrier == "AA"],
    pch = 23,
    col = "red",
    main = "AA", 
    xlab = "Arr_Delay", 
    ylab = "Dep_Delay"
    )
```

```r
par(mfrow = c(1,2))
plot(
    x = tibb$arr_delay[tibb$carrier == "UA"], 
    y = tibb$dep_delay[tibb$carrier == "UA"], 
    pch = 17, 
    main = "UA", 
    xlab = "Arr_Delay", 
    ylab = "Dep_Delay",
    col = "purple",
    xlim = c(0, 1000),
    ylim = c(0, 1000)
    )
plot(
    x = tibb$arr_delay[tibb$carrier == "AA"],
    y = tibb$dep_delay[tibb$carrier == "AA"],
    pch = 23,
    col = "red",
    main = "AA", 
    xlab = "Arr_Delay", 
    ylab = "Dep_Delay",
    xlim = c(0, 1000),
    ylim = c(0, 1000)
    )
```

```r
par(mfrow = c(1,1))
plot(
    x = tibb$arr_delay[tibb$carrier == "UA"], 
    y = tibb$dep_delay[tibb$carrier == "UA"], 
    pch = 17, 
    main = "Arrival vs Departure Delay", 
    xlab = "Arr_Delay",
    ylab = "Dep_Delay",
    col = "blue"
    )
text(100, 400, label = "United Airways", adj =  1)
mtext(text = "in minutes", side = 2, adj = 1)
abline(h=250)
abline(v=250)
abline(h=250)
abline(100,1)
```

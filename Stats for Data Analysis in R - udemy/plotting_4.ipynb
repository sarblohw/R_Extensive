```r
install.packages("pacman")
pacman::p_load("nycflights13")
```

```r
tibb <- nycflights13::flights
```

```r
par(mfrow = c(1,1))
plot(
    x = tibb$arr_delay[tibb$carrier == "UA"], 
    y = tibb$dep_delay[tibb$carrier == "UA"], 
    pch = 17, 
    main = "Arrival vs Departure Delay", 
    xlab = "Arr_Delay",
    ylab = "Dep_Delay",
    col = "blue",
    cex = 0.5,
    col.main = "red",
    font.main = 4,
    col.lab = "red",
    font.lab = 3
    )
text(100, 400, label = "United Airways", adj =  1, col = "green", font = 2, cex = 1.2)
mtext(text = "in minutes", side = 2, adj = 1)
abline(h=250)
abline(v=250)
abline(h=250)
abline(100,1)
```
